```python
import pandas as pd
```

```python
df = pd.read_csv("monte_monte_2.csv")
```

```python
df.describe()
```

```python
set(df.ActionDirection)
```

```python
df[['GameMe0', 'GameMe1', 'ActionIndex']]
```

```python
X = df[['GameMe0', 'GameMe1']].as_matrix()
```

```python
Y = df.ActionIndex
```

```python
df['TargetPos'] = [X.item(pos) for pos in enumerate(Y)]
```

```python
from sklearn.linear_model import LogisticRegression
m = LogisticRegression()
df['ActionDirectionAsInt'] = df.ActionDirection.map({'-1': 0, '-4': 1, '1': 2, '4': 3, 'WIN': 4})
xs = df[['TargetPos', 'ActionDirectionAsInt']]
ys = df.ActionIndex
m.fit(xs[::2], ys[::2])
m.score(xs[1::2], ys[1::2])
```

```python
from sklearn.svm import SVC
m = SVC()
m.fit(xs[::200], ys[::200])
m.score(xs[1::200], ys[1::200])
```

```python
df['TheOtherPos'] = df.GameMe0 + df.GameMe1 - df.TargetPos
```

```python
xs = df[['TargetPos', 'ActionDirectionAsInt', 'TheOtherPos']]
m = LogisticRegression()
m.fit(xs[::2], ys[::2])
m.score(xs[1::2], ys[1::2])
```

```python
m = SVC()
m.fit(xs[::200], ys[::200])
m.score(xs[1::200], ys[1::200])
```

```python
df['TargetPosX'] = df.TargetPos % 4
df['TargetPosY'] = df.TargetPos // 4
df['TheOtherPosX'] = df.TheOtherPos % 4
df['TheOtherPosY'] = df.TheOtherPos // 4
df['TargetPosStr'] = df.TargetPos.astype(str)
df['TheOtherPosStr'] = df.TheOtherPos.astype(str)

xs = df[['TargetPosX', 'TargetPosY', 'TargetPosStr', 'ActionDirection', 'TheOtherPosX', 'TheOtherPosY', 'TheOtherPosStr']]
xs = pd.get_dummies(xs)
xs.head()
```

```python
m = LogisticRegression()
m.fit(xs[::2], ys[::2])
m.score(xs[1::2], ys[1::2])
```

```python
list(sorted(zip(m.coef_[0], xs.columns)))
```
